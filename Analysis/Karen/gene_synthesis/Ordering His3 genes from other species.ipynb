```python
initially_optimized_DNA_seq_file = 'His3 sequences from different species - DNA sequences preoptimized with CodonOpt.fa'
initially_optimized_DNA_seq = read_fasta(initially_optimized_DNA_seq_file)

common_codons = get_common_codons(codon_usage, ['Saccharomyces_cerevisiae'])
unwanted_renzymes = ['BpiI', 'BsaI']#, 'BtgZI', 'BsmFI', 
#                'BstBI', 'SpeI',  'AvrII', 'PvuII', 'XhoI', 'KpnI', 'SacII', 'XbaI', 'SalI', 'BsmFI', 
#                'EcoRI', 'BamHI', 'HindIII', 'AgeI', 'NotI']
```

```python
Gibson_fwd = 'GCAGGCAAGATAAACGAAGGCAAAG' #3'
Gibson_rev = 'TGACACCGATTATTTAAAgctgcag' #3'
```

```python
BpiI_5 = 'GAAGACac'
BpiI_3 = 'acGTCTTC'
```

```python
genes_to_order = []
gene_names_to_order = []
for sr in initially_optimized_DNA_seq:
    re_sites_removed = remove_restriction_sites_silently_for_several_enzymes(str(sr.seq), unwanted_renzymes)
    optimized = optimize_seq(re_sites_removed, common_codons)
    re_sites_removed = remove_restriction_sites_silently_for_several_enzymes(optimized, unwanted_renzymes)
    with_gibson_arms = Gibson_fwd + re_sites_removed + Gibson_rev
    with_restriction_flanks = BpiI_5 + with_gibson_arms + BpiI_3
    genes_to_order.append(with_restriction_flanks)
    gene_names_to_order.append(sr.description)
```
