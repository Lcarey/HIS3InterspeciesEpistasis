```python
def get_mutation_from_Scer(permutation_lucas, position_lucas):
    position_lucas = int(position_lucas)
    wt_state = Scer_Uniprot[position_lucas-1]
    if wt_state not in permutation_lucas or permutation_lucas[-1] == wt_state:
        return None
    else:
        return str(position_lucas) + ''.join([char for char in permutation_lucas if char != wt_state])

def get_mutation_from_Scer3(permutation_lucas, position_lucas):
    position_lucas = int(position_lucas)
    wt_state = Scer_Uniprot[position_lucas-1]
    if wt_state not in permutation_lucas:
        return None
    else:
        return wt_state

def get_mutation_from_Scer2(permutation_lucas, position_lucas):
    position_lucas = int(position_lucas)
    wt_state = Scer_Uniprot[position_lucas-1]
    return str(position_lucas) + permutation_lucas[0]

lucas_subset = lucas_sign_epistasis[(lucas_sign_epistasis.pBon < 0.01) & (lucas_sign_epistasis.SegN == 5) & \
                                    (lucas_sign_epistasis.VarPos_absolute + lucas_sign_epistasis.SubPos_absolute > 0)].copy()

lucas_subset['mutation_Var'] = lucas_subset.apply(lambda df: get_mutation_from_Scer2(df['Perm'], df['VarPos_absolute']), axis=1)
lucas_subset['mutation_Sub'] = lucas_subset.apply(lambda df: get_mutation_from_Scer2(df['SubPerm'], df['SubPos_absolute']), axis=1)

lucas_subset['mutation_Var_wt'] = lucas_subset.apply(lambda df: get_mutation_from_Scer3(df['Perm'], df['VarPos_absolute']), axis=1)
lucas_subset['mutation_Sub_wt'] = lucas_subset.apply(lambda df: get_mutation_from_Scer3(df['SubPerm'], df['SubPos_absolute']), axis=1)


lucas_subset['mutation_pair'] = lucas_subset.apply(lambda df: (df['mutation_Var'], df['mutation_Sub']), axis=1)
lucas_subset.sort_values('logodds', ascending=False)[['VarPos_absolute', 'SubPos_absolute', u'Perm', 'SegN',
                                                      'SubPerm', 'logodds', 'mutation_Var', 'mutation_Sub', 'mutation_pair', 
                                                     'mutation_Var_wt', 'mutation_Sub_wt']]
```
